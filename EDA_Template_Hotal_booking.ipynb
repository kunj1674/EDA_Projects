#### Duplicate Values

````python
# Dataset Duplicate Value Count
df.duplicated().sum()
````

````python
# Dataset Duplicate Value Count in each row
for column in column_list:
  print(column,": ",df.duplicated(subset=[column]).sum())
````

#### Missing Values/Null Values

````python
# Missing Values/Null Values Count
df.isnull().sum()
````

````python
# Visualizing the missing values
Nullvalue_list=[]
for column in column_list:
  Nullvalue_list.append(df[column].isnull().sum())

Nullvalue_list
````

````python
plt.figure(figsize=(21,3))
plt.bar(x=column_list,height=Nullvalue_list)
plt.title("Number of null values in each column")
plt.xlabel("Null value count of each column")
plt.xticks(rotation=90)
plt.ylabel("Column Name")
````
