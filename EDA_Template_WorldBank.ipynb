 **Project Summary: World Bank Global EdStats Analysis**

***Overview:***

This project involves a comprehensive Exploratory Data Analysis (EDA) on the World Bank’s Global Education Statistics (EdStats) dataset. The primary goal is to uncover patterns in how different countries and regions manage, report, and categorize their socio-economic and agricultural data. By examining variables such as income groups, regional locations, and the recency of national surveys, this analysis provides a clear picture of the global data landscape, highlighting the gap between developed and developing nations.

***The Objective:***

The core business objective is to identify which regions lack up-to-date information and how economic status affects a country’s ability to maintain modern data systems. For an organization like the World Bank or a global educational NGO, these insights are crucial for deciding where to allocate resources, technical support, and funding to improve global transparency and reporting standards.

***Data Cleaning and Preparation:***

The raw dataset consisted of 241 country-level records with 32 distinct features. Initial inspection revealed significant challenges: several columns were entirely empty (like Unnamed: 31), and many others had high percentages of missing values.

To ensure the analysis was reliable, I performed Data Wrangling which included:

Removing redundant or completely empty columns.

Handling missing values using Median Imputation for numerical columns (like trade and industrial data) to avoid the influence of outliers.

Applying Mode Imputation for categorical data to maintain the consistency of the dataset.

Formatting currency and regional labels to ensure uniformity across all visualizations.

***Key Insights from Analysis:***

Through a combination of Univariate, Bivariate, and Multivariate visualizations, several critical trends emerged:

Economic Disparity: The analysis of Income Groups via pie charts showed a dominant share of "High Income" and "Upper Middle Income" countries. However, when cross-referenced with regions, it became clear that "Low Income" countries are heavily concentrated in Sub-Saharan Africa and South Asia.

Data Recency Gaps: A major finding was the "Recency Gap." High-income nations consistently have trade and population census data from 2010–2012. In contrast, many lower-income nations are still relying on census data from the late 1990s or early 2000s, which significantly hinders accurate policymaking.

Agricultural Focus: By analyzing agricultural census data, the project identified that regions like Sub-Saharan Africa and Latin America have a higher frequency of farming-related data collections compared to their industrial reporting. This highlights the agricultural dependency of these economies.

Financial Infrastructure: The heatmap comparing Lending Categories to Income Groups revealed that "IDA" (International Development Association) lending is almost exclusively tied to low-income regions, confirming that financial aid is being directed toward the most vulnerable economies.

***Conclusion and Business Impact:***

The project concludes that while the world is becoming more data-driven, a "Digital and Statistical Divide" remains. For businesses or international bodies, this means that investment in Low Income regions must first begin with improving data infrastructure.

The insights gained here allow stakeholders to move away from a "one-size-fits-all" approach. Instead, they can target specific regions (like those using outdated PPP survey years) for specialized technical assistance. Ultimately, this EDA serves as a roadmap for improving global data quality, ensuring that no region is left behind in the pursuit of evidence-based development.

### Variables Description

Country Code: A unique 3-letter international code for each country or region.

Short Name: The common name/short name of the country.

Table Name: The name of the country as it appears in standard World Bank data tables.

Long Name: The official ceremonial name of the country.

2-alpha code: A 2-letter country code.

WB-2 code: The internal 2-letter code used by the World Bank.

Region: The geographical area where the country is located .

Income Group: Classification based on Income category.

Currency Unit: The official currency used in the country.

Other groups: Indicates if the country belongs to specific economic or geographic aggregates.

National accounts base year: The year used as a benchmark for calculating constant price economic data.

National accounts reference year: The year used for price comparisons in economic volume measures.

SNA price valuation: Specifies whether the System of National Accounts uses basic prices or producer prices.

System of National Accounts: The standard framework used by the country to compile economic activity.

Alternative conversion factor: Used when the official exchange rate is deemed unrepresentative for converting local currency to USD.

PPP survey year: The year the Purchasing Power Parity survey was conducted.

Government Accounting concept: Defines how the government records its financial transactions.

Special Notes: Qualitative details regarding data collection, adjustments, or specific historical contexts for that country’s data.

Lending category: Classification for World Bank borrowing eligibility (IDA, IBRD, or Blend).

Balance of Payments Manual in use: Indicates if the country follows BPM5 or BPM6 standards for international transactions.

External debt Reporting status: Status of the country's reporting to the World Bank’s Debtor Reporting System.

System of trade: Indicates if the country uses a General or Special trade system for imports/exports.

IMF data dissemination standard: The standard (SDDS or GDDS) the country follows for releasing economic and financial data to the public.

Latest population census: The year the most recent full population count was conducted.

Latest household survey: The most recent major survey (like DHS or MICS) used to collect health and social data.

Source of most recent Income and expenditure data: The specific survey used to track household wealth and poverty.

Vital registration complete: Indicates whether the country's birth and death registration systems are considered 90% or more complete.

Latest agricultural data: These columns track the most recent year for which agricultural data is available.

Latest industrial data: These columns track the most recent year for which industrial data is available.

Latest trade data: These columns track the most recent year for which trade data is available.

Latest water data: These columns track the most recent year for which water data is available.

Latest Withdrawal data: These columns track the most recent year for which withdrawal data is available.
